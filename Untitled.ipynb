```python
import pandas as pd 
import numpy as np 
```

```python
match=pd.read_csv('ipl data set/matches.csv')
delevry=pd.read_csv('ipl data set/deliveries.csv')
```

```python
delevry.head(3)
```

```python
total_score_df=delevry.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
total_score_df
```

```python
total_score_df=total_score_df[total_score_df['inning']==1]
```

```python
total_score_df
```

```python
match_df=match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df['team1'].unique()
```

```python
teams=[
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
delevry_df=match_df.merge(delevry,on='match_id')
```

```python
delevry_df=delevry_df[delevry_df['inning']==2]
```

```python
delevry_df
```

```python
delevry['total_runs'] = pd.to_numeric(delevry['total_runs'], errors='coerce')
```

```python
delevry_df['curr_score'] = delevry_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delevry_df
```

```python
delevry_df['runs_left']=delevry_df['total_runs_x']-delevry_df['curr_score']
```

```python
delevry_df
```

```python
delevry_df['balls_left']=126-(delevry_df['over']*6+delevry_df['ball'])
```

```python
delevry_df
```

```python
delevry_df.head()
```

```python
# Step 1: player_dismissed numeric 0/1
delevry_df['player_dismissed'] = delevry_df['player_dismissed'].fillna('0')
delevry_df['player_dismissed'] = delevry_df['player_dismissed'].apply(lambda x: 0 if x=='0' else 1)
delevry_df['player_dismissed'] = delevry_df['player_dismissed'].astype('int')

# Step 2: cumulative wickets per match and per inning (important!)
delevry_df['cumulative_wickets'] = delevry_df.groupby(['match_id','inning'])['player_dismissed'].cumsum()

# Step 3: wickets left (max 10 per inning)
delevry_df['wickets_left'] = 10 - delevry_df['cumulative_wickets']

# Step 4: prevent negative values in case of data errors
delevry_df['wickets_left'] = delevry_df['wickets_left'].clip(lower=0)
```

```python
delevry_df.tail()
```

```python
delevry_df['curr_runrate']=(delevry_df['curr_score']*6)/(120-delevry_df['balls_left'])
```

```python
delevry_df['curr_runrate']
```

```python
delevry_df['req_runrate']=(delevry_df['runs_left']*6)/(delevry_df['balls_left'])
```

```python
delevry_df['req_runrate']
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
delevry_df['result']=delevry_df.apply(result,axis=1)
```

```python
delevry_df.columns
```

```python
final_df=delevry_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','curr_runrate','req_runrate','result']]
```

```python
final_df
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=final_df.drop(['result'],axis=1)
Y=final_df['result']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
   ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team','bowling_team','city'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
import pickle
pickle.dump(pipe, open('ipl_result.pkl', 'wb'))
```
